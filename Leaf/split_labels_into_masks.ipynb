```python
import sys 
sys.executable
from plantcv import plantcv as pcv
import numpy as np
#np.version.version
```

```python
pcv.params.debug = "plot"
```

```python
img, img_path, img_filename = pcv.readimage("plant00000_label.png")
print(img.shape)
print(img.reshape(-1, img.shape[2]).shape)
```

```python
uniq_rgb = np.unique(img.reshape(-1, img.shape[2]), axis=0)
uniq_rgb = np.delete(uniq_rgb, 0, 0)
```

```python
np.shape(uniq_rgb)
```

```python
#
```

```python
for rgb in uniq_rgb:
    mask = np.zeros(np.shape(img)[:2], dtype=np.uint8)
    mask[np.all(img == rgb, axis=-1)] = 255
    mask = mask.astype(np.bool)
    pcv.plot_image(mask)
```
